Exercise 2: The "Batch" Processor
- Task: You have a batch of 32 grayscale images, each $28 \times 28$. Create this tensor. Then, "flatten" it so each image is just a single long line of numbers (vector).
- Goal: Learn how models "see" images as flat input vectors.

```python
# batch of 32 images, 1 is the channel, and 28x28 pixels
batch = torch.rand(32, 1, 28, 28)
# torch.Size([32, 1, 28, 28])
batch.shape

# Transform 28x28 into 784 using flatten
torch.flatten(batch, start_dim=1).shape

# Transform 28x28 into 784 using view
batch.view(32, -1).shape
```
