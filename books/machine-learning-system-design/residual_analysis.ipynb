```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Generate non-linear data (a simple curve)
np.random.seed(42)
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = X**2 + np.random.normal(0, 5, 100).reshape(-1, 1) # Quadratic relationship

# 2. Fit a Linear Regression model (The "Wrong" Model)
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

# 3. Calculate Residuals
residuals = y - predictions

# 4. Visualization
plt.figure(figsize=(10, 5))

# Residual Plot
plt.scatter(predictions, residuals, color='purple', alpha=0.6)
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Residual Plot: Linear Model on Quadratic Data')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals (Actual - Predicted)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generate data with constant variance (Random Scatter)
X_rand = np.linspace(0, 100, 100).reshape(-1, 1)
y_rand = 2.5 * X_rand + np.random.normal(0, 10, (100, 1)) # Constant noise

model_rand = LinearRegression().fit(X_rand, y_rand)
res_rand = y_rand - model_rand.predict(X_rand)

plt.scatter(model_rand.predict(X_rand), res_rand, color='blue', alpha=0.6)
plt.axhline(0, color='red', linestyle='--')
plt.title("Ideal: Random Scatter")
plt.show()
```

```python
# Generate data with increasing variance (Funnel Shape)
X_funnel = np.linspace(1, 100, 100).reshape(-1, 1)
# Noise grows as X grows
noise = np.random.normal(0, 1, (100, 1)) * (X_funnel * 0.5)
y_funnel = 1.5 * X_funnel + noise

model_funnel = LinearRegression().fit(X_funnel, y_funnel)
res_funnel = y_funnel - model_funnel.predict(X_funnel)

plt.scatter(model_funnel.predict(X_funnel), res_funnel, color='orange', alpha=0.6)
plt.axhline(0, color='red', linestyle='--')
plt.title("Warning: Funnel Shape (Heteroscedasticity)")
plt.show()
```
